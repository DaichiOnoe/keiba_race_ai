# 既存のhorseデータに対して新規のものを新しく付け加えていくコードを作成

```python
horse_results_old["horse_id"].isin(results_horse["horse_id"]).sum()
```

```python
set(horse_results_old["horse_id"]) & set(results_horse["horse_id"])
```

```python
# 関数化
def updating_rawdf(old_df, new_df):
    df = pd.concat([
        old_df[~old_df["horse_id"].isin(new_df["horse_id"])], #新規のhorse_idデータを除外した既存のhorse_idデータ
        new_df, #新規のデータ
    ])
    return df
```

```python
df_updated = updating_rawdf(horse_results_old, results_horse)
df_updated
```

```python
# 数チェック
len(horse_results_old), len(results_horse), len(df_updated)
```

# 血統データをスクレイピング

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import requests
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas as pd
import re
url = "https://db.netkeiba.com/horse/ped/2020103575/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
        }
# Requestオブジェクトを作成
req = Request(url, headers=headers)
        # リクエストを送信 # URLのHTMLを取得 #スクレイピング
html = urlopen(req).read()
```

```python
pd.read_html(html)[0]
```

```python
soup = BeautifulSoup(html, "html.parser")
soup = soup.find("table", class_="blood_table detail")
soup
```

```python
td_list = soup.find_all("td")
len(td_list)
```

```python
td_list[0].find("a")["href"]
```

```python
re.findall(r"horse/(\d+)/", td_list[0].find("a")["href"])[0]
```

```python
ped_id_list = []
for td in td_list:
        try:
            ped_id = re.findall(r"horse/(\w+)/", td.find("a")["href"])[0]
        except:
            ped_id = None
        ped_id_list.append(ped_id)
```

```python
len(ped_id_list)
```

```python
# .add_prefix("ped_")で列名に"ped_"を付ける
ped = pd.DataFrame(ped_id_list).T.add_prefix("ped_")
ped
```
