# 既存のhorseデータに対して新規のものを新しく付け加えていくコードを作成

```python
horse_results_old["horse_id"].isin(results_horse["horse_id"]).sum()
```

```python
set(horse_results_old["horse_id"]) & set(results_horse["horse_id"])
```

```python
# 関数化
def updating_rawdf(old_df, new_df):
    df = pd.concat([
        old_df[~old_df["horse_id"].isin(new_df["horse_id"])], #新規のhorse_idデータを除外した既存のhorse_idデータ
        new_df, #新規のデータ
    ])
    return df
```

```python
df_updated = updating_rawdf(horse_results_old, results_horse)
df_updated
```

```python
# 数チェック
len(horse_results_old), len(results_horse), len(df_updated)
```

```python

len(horse_results_old) + len(results_horse) 
```

```python
# 重複がないかチェック
df_updated.duplicated(subset=["horse_id", "日付"]).sum()
```

```python
# 新しく取得したhorse_idのデータないhorse_idのデータに変更はないはず
len(horse_results_old[~horse_results_old["horse_id"].isin(results_horse["horse_id"])]) ,len(df_updated[~df_updated["horse_id"].isin(results_horse["horse_id"])])
```
